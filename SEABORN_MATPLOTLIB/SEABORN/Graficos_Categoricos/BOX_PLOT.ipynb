```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head()
```

```python
sns.boxplot(data=mart,x='Total', width = 0.5)
```

```python
sns.boxplot(data=mart, y ='Payment',x='Total', width = 0.5, hue = 'Gender')
```

## estilizar e adicionar marcador para a media

```python
sns.boxplot(data=mart, y ='Payment',x='Total', showmeans = True,hue = 'Gender',
            meanprops={'marker':'o' ,
                       'markerfacecolor': 'white' , 
                        'markersize': '10',
                        'markeredgecolor': 'black'} )
```

## linewidth = 5 
### borda ao redor do boxplot

```python
sns.boxplot(data=mart, y ='Payment',x='Total', showmeans = True,hue = 'Gender',
            meanprops={'marker':'o' ,
                       'markerfacecolor': 'white' , 
                        'markersize': '10',
                        'markeredgecolor': 'black'},linewidth = 5 )
```
