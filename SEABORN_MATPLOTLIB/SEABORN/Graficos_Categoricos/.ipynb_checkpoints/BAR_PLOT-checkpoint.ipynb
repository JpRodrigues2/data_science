```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart  = pd.read_excel('supermarket_sales.xlsx')
mart.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data = mart, x = 'Product line', y = 'Total', hue = 'Product line')
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data = mart, x = 'Product line', y = 'Total', hue = 'City')
```

## Ordenar as barras 

```python
X = mart['Product line'].sort_values()
X.unique()

plt.figure(figsize=(12,7))
sns.barplot(data = mart, x = 'Product line', y = 'Total', hue = 'City', order = ['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], hue_order = ['Mandalay', 'Yangon','Naypyitaw'])
```

```python
plt.figure(figsize=(12,7))
sns.barplot(data = mart, x = 'Product line',
                         y = 'Total', hue = 'City',
                        order = ['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],
                        hue_order = ['Mandalay', 'Yangon','Naypyitaw'],
                        capsize = 0.2
           )
```

## Remove barra 

```python
plt.figure(figsize=(12,7))
sns.barplot(data = mart, x = 'Product line',
                         y = 'Total', hue = 'City',
                        order = ['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],
                        errorbar = None)
```

## Palette

```python
plt.figure(figsize=(12,7))
sns.barplot(data = mart, x = 'Product line',
                         y = 'Total', hue = 'City',
                        order = ['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],
                        palette ='cividis')
```

## Saturation

```python
plt.figure(figsize=(12,7))
sns.barplot(data = mart, x = 'Product line',
                         y = 'Total', hue = 'City',
                        order = ['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],
                        palette ='cividis',
                        saturation = 0.1)
```
