```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head()
```

## CATPLOT --> Cria subgráficos com base nas variáveis que você escolhe, usando os parâmetros col e row.

```python
sns.catplot(data = mart, y = 'Total' ,x = 'Payment', kind='violin', col='Gender', row = 'Customer type', hue= 'Payment')
```
