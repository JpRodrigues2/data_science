```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart
```

## GrÃ¡fico de barras vertical

```python
plt.figure(figsize=(16,8))
sns.countplot(data = mart, x = 'Product line', hue = 'Gender')

# HUE --> nesse caso divide a coluna em subgrupos e coloca cores diferentes
```

```python
plt.figure(figsize=(16,8))
sns.countplot(data = mart, x = 'Product line', hue = 'Product line')
```
