## O Swarm Plot
### sns.swarmplot() --> é um gráfico semelhante ao Strip Plot, mas com uma diferença fundamental: os pontos nunca se sobrepõem. Em vez disso, os pontos são distribuídos como um enxame de abelhas (swarm), garantindo que cada valor fique visível.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head()
```

```python
sns.swarmplot(data = mart, y = 'Total')
```

```python
sns.swarmplot(data = mart, y = 'Total' ,x = 'Payment', hue= 'Gender')
```

```python
sns.swarmplot(data = mart, y = 'Total' ,x = 'Payment', hue= 'Gender', dodge = True)
```

```python
sns.swarmplot(data = mart, y = 'Total' ,x = 'Payment',marker='d',size=5)

# SIZE --> tamanho dos icones (markers)
```

```python
sns.swarmplot(data = mart, y = 'Total' ,x = 'Payment', color='black')
sns.boxplot(data = mart, y = 'Total' ,x = 'Payment')
```

```python
sns.swarmplot(data = mart, y = 'Total' ,x = 'Payment', color='black')
sns.violinplot(data = mart, y = 'Total' ,x = 'Payment')
```
