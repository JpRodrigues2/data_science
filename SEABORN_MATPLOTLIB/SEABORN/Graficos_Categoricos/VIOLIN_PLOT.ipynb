```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head()
```

```python
sns.violinplot(data = mart, y = 'Total')
```

```python
sns.violinplot(data = mart, y = 'Total', x = 'Payment', hue = 'Payment')
```

## SPLIT --> DUPLICA OU NAO OS VIOLIN'S , True deixa meio a meio
## INNER --> muda o estilo que contem as informacoes de mediana e quartis (box, None, quartile)
## BW_METHOD --> controla a suavização do violino. Valores menores tornam o gráfico mais detalhado.

```python
sns.violinplot(data = mart, y = 'Total', x = 'Payment', hue = 'Gender', split = True, inner = 'quartile', bw_method = 0.2)
```

## CUT --> Controla a extensão das bordas do violin podendo extrapolar os valores reais , default = 2 

```python
sns.violinplot(data = mart, y = 'Total', x = 'Payment', hue = 'Gender', split = True, inner = 'quartile', bw_method = 0.2, cut = 0, palette = 'cividis')
```
