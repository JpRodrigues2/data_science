```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head()
```

```python
sns.stripplot( data = mart, y ='Total',  jitter = 0.2)  

# JITTER --> Espalha os pontos para melhor vizualização
```

```python
sns.stripplot( data = mart, y ='Total',  jitter = 0.2, linewidth = 0.5)  


# LINEWIDTH --> Contorno em volta dos pontos
```

# Retorna 3 strip plot, baseado no 'Total', e divide por 'Gender'

```python
plt.figure(figsize=(15,10))
sns.stripplot( data = mart, y ='Total', x = 'Payment', hue = 'Gender', linewidth = 0.5, dodge = False)  
```

```python
plt.figure(figsize=(15,10))
sns.stripplot( data = mart, y ='Total', x = 'Payment', hue = 'Gender', linewidth = 0.5, dodge = True)  

# DODGE --> Separa os pontos sobrepostos, pontos masculinos e femininos ficam lado a lado.
```

```python
sns.stripplot( data = mart, y ='Total', x = 'Payment', jitter = 0.1, hue = 'Payment')
sns.violinplot( data = mart, y= 'Total', x = 'Payment', color = 'gray')

plt.show()
```

```python
sns.stripplot( data = mart, y ='Total', x = 'Payment', jitter = 0.1, color = 'black')
sns.boxplot(data=mart, y ='Total', x='Payment', color = 'gray')
```
