```python
import pandas as pd
import numpy as np
df = pd.read_csv('bank.csv' , encoding = 'UTF-8' , sep = ';')
```

```python
df = pd.DataFrame(df)

df.iloc[0 , df.columns.get_loc('marital')] = np.nan 

#TRANFORMA A LINHA 0 , COLUNA 'MARITAL' EM NaN
```
