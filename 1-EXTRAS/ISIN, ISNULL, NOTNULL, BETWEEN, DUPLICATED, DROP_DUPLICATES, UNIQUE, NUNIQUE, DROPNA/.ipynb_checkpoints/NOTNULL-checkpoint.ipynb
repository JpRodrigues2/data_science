# NOTNULL

### Contr√°rio do ISNULL, NOTNULL retorna valores presentes no dataframe

```python
import pandas as pd
import numpy as np
df = pd.read_csv('bank.csv' , encoding = 'UTF-8' , sep = ';')
```

```python
df.head()
```

```python
np.random.seed(1)
arr = np.random.randint(0,100,50)
arr
```

```python
df.iloc[arr , df.columns.get_loc('marital')] = np.nan
```

```python
display(df[df['marital'].notnull()]['marital'].head())

#RETORNA O QUE EXISTE NO DATAFRAME (~NaN)
```

```python
display(df[df['marital'].notnull()].head(10))
```
