```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as srn
```

### Dados Normais

```python
dados = stats.norm.rvs(size = 1000) # Gerando números aleatórios da distribuição normal
```

```python
#srn.histplot(dados,kde = True)

plt.hist(dados, bins = 20) #passe a base de dados, e a quantidade de intervalos
plt.title('Dados')
```

```python
#Gráfico para Verificar se a distribuição é normal
#gráfico QQ-Plot 
fid, ax = plt.subplots()
stats.probplot(dados,  fit = True ,plot = ax)
plt.show
```

```python
# teste de Shapiro-Wilk  , caso p < 0.05 , Não está normalmente distribuído
stats.shapiro(dados)
```
