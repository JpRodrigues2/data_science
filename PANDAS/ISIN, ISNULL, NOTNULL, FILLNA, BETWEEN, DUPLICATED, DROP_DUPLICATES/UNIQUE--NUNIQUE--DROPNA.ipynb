```python
import pandas as pd
import numpy as np
df = pd.read_csv('bank.csv' , encoding = 'UTF-8' , sep = ';')
```

```python
df.head()
df.iloc[0 , df.columns.get_loc('marital')] = np.nan
```

## UNIQUE - Retorna um array np com os valores únicos

```python
df['marital'].unique()
```

## NUNIQUE - Retorna a quantidade de valores únicos em uma Series ou DataFrame.

```python
df['marital'].nunique()
```

```python
df['marital'].nunique(dropna = False) 

#DROPNA --> FALSE, Considera o NaN como valor 
```

```python
df.head()
```

# DROPNA


```python
df.dropna()
df.head()
```

### Remover apenas se todas as colunas forem NaN 

df.dropna(how='all').head()

###  Remover linhas que tenham NaN apenas em colunas específicas

```python
df.dropna(subset=['marital']).head()
```

### Remover linhas que tenham pelo menos N valores não nulos

```python
df.dropna(thresh=2).head()
```
