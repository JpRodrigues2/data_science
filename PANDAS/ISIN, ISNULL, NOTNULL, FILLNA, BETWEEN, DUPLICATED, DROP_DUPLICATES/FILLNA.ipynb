### Substitui valores NaN 

```python
import pandas as pd
import numpy as np
df = pd.read_csv('bank.csv' , encoding = 'UTF-8' , sep = ';')

np.random.seed(1)
arr = np.random.randint(0,100,50)
arr

df.iloc[arr , df.columns.get_loc('marital')] = np.nan
```

```python
x = df[df['marital'].isna()]
x.head()
```

```python
df['marital'] = df['marital'].fillna('teste')  # Preenche os valore NaN com o parametro 'teste'
df.head()  
```
