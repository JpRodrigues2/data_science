```python
import pandas as pd
import numpy as np

df = pd.read_excel('reshaping.xlsx')
df
```

# PIVOT
### df.pivot(index, columns, values)


Index --> coluna que irá servir como indice \
Columns --> colunas que desejam ser apresentadas \
Values --> informação final que deve ser apresentada

```python
df_pivot = df.pivot(index = 'Estudante' , columns = ['Categoria', 'Semana', 'Tópico'] , values = 'Nota')

df_pivot
```

# STACK 

### A função stack() transforma colunas em um MultiIndex de linhas, reorganizando os dados.

```python
df_pivot.stack(level = 0, future_stack = True)

#LEVEL --> Colunas , começa no 0 e vai descendo
```

# UNSTACK
### O CONTRÁRIO DO STACK

```python
df_pivot.stack(level = 0, future_stack = True).unstack(0)

#Level --> tambem começando no 0 e seguindo para a direita 
```
