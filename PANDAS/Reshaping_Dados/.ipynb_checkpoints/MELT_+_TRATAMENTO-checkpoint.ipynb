```python
import pandas as pd

dados = pd.read_csv('dados_satisfacao.csv')

dados
```

# MELT
### df.melt(id_vars, value_vars, var_name, value_name)


```python
df1 = dados.melt( 
    id_vars=['cliente_id','idade','genero'],
    value_vars = [
        'satisfacao_geral_01-2024',
        'satisfacao_geral_02-2024',
        'satisfacao_geral_03-2024',
        'satisfacao_produto_01-2024',
        'satisfacao_produto_02-2024',
        'satisfacao_produto_03-2024',
        'satisfacao_servico_01-2024',
        'satisfacao_servico_02-2024',
        'satisfacao_servico_03-2024'],
    var_name = 'categoria_mes',
    value_name = 'nota'
)
```

```python
df1[['categoria','mes']] = df1['categoria_mes'].str.rsplit('_', n=1, expand = True)

# RSPLIT --> split que le da direita para esquerda
# N-1 --> dividir a string em duas partes: a parte antes do último _ e a parte depois.
# EXPAND = True: Isso faz com que o resultado seja retornado como um DataFrame, com uma nova coluna para cada parte da divisão.
```

```python
df1.drop('categoria_mes', axis = 1, inplace = True)
```

```python
df1
```

```python
df1['mes'] = pd.to_datetime(df1['mes'], format = '%m-%Y')
```

```python
df1 
```

```python
df1['mes'] = df1['mes'].dt.month
```
