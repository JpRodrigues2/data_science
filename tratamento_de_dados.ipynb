```python
import pandas as pd
import statistics as sts
```

```python
df = pd.read_csv("tempo.csv" , sep = ';')
display(df)
```

```python
med_temp = sts.median(df["Temperatura"])
med_temp
```

```python
df.loc[df['Temperatura'] > 100, 'Temperatura'] = med_temp
display(df)
```

```python
med_umi = sts.median(df["Umidade"])
med_umi
```

```python
df.loc[df["Umidade"] > 100, 'Umidade'] = med_umi
display(df)
```

```python
df['Umidade'].fillna(med_umi, inplace = True)
```
