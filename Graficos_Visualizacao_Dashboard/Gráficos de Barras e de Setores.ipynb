```python
import pandas as pd
```

```python
base = pd.read_csv('insect.csv')
base.head()
```

```python
agrupado = base.groupby('spray').sum() #groupby é para contar, e o .sum é a operação
agrupado
```

```python
agrupado.plot.bar(color = 'blue') #passa a base de dados (agrupado), .plot (gráfico), .bar (barras)
```

```python
agrupado.plot.pie(subplots = True, legend = True) #subplots = True é para gerar 2 gráficos
                                                  # legend é para haabilitar legenda
```
