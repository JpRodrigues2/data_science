```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
h = np.histogram(base.iloc[:,1], bins = 6) #bins divide os intervalos no histograma
h
```

```python
h = np.histogram(base.iloc[:,1], bins = 6) #bins divide os intervalos no histograma

plt.hist(base.iloc[:,1],bins = 6) #gera o histograma
plt.title("Árvores")
plt.ylabel("Frequência")
plt.xlabel('Altura')
```

```python
srn.histplot(base.iloc[:,1], kde = False, bins = 6, color = 'blue').set(title="Árvores") 
# kde é linha de densidade
```

```python
srn.kdeplot(base.iloc[:,1], color="blue").set(title = "Árvores")
```

```python
srn.histplot(base.iloc[:,1], kde = False, bins = 6, color = 'blue').set(title="Árvores")
# linha de densidade mostra as diferenças de forma menos abruptas
```
