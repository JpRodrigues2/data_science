```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
base = pd.read_csv('co2.csv')
display(base)
```

```python
srn.scatterplot(x = base.conc , y = base.uptake, hue = base.Type)
#hue atribui cores diferentes para 'Type'
```

```python
m = base.loc[base['Type'] == 'Mississippi']
q = base.loc[base['Type'] == 'Quebec']
```

```python
fig, axes = plt.subplots(1,2)

srn.scatterplot(x = q.conc, y = q.uptake, ax=axes[0]).set(title="Quebec")
srn.scatterplot(x = m.conc, y = m.uptake, ax=axes[1]).set(title="Mississippi")

plt.tight_layout()
```

```python
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']

fig, axes = plt.subplots(1,2 )
                        
srn.scatterplot(x = ch.conc,y=  ch.uptake, ax = axes[0])
srn.scatterplot(x = nc.conc,y= nc.uptake, ax = axes[1])

plt.tight_layout()
```
