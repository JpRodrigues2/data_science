```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('trees.csv')
```

```python
plt.scatter (x = base.Girth, y = base.Volume, color = 'blue', facecolors ='none', marker = '*') #grafico de dispersão
# no seaborn o metodo scatter precisa que seja passado x e y

plt.title("Árvores")
plt.ylabel("Circunferência")
plt.xlabel('Volume')
```

```python
plt.plot(base.Girth, base.Volume)   #gráfico comum
plt.title("Árvores") 
plt.ylabel("Circunferência")
plt.xlabel('Volume')
```

```python
sns.regplot(x = base.Girth, y = base.Volume, data = base, x_jitter = 0.3, fit_reg = False)
```

```python
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(x = base.Girth, y = base.Volume, color="green", marker="*")

plt.subplot(2,2,2)
plt.scatter(x = base.Girth, y = base.Height)

plt.subplot(2,2,3)
plt.scatter(x = base.Height, y = base.Volume)

plt.subplot(2,2,4)
plt.hist(base.Volume)
```
