```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
base2 = pd.read_csv('chicken.csv')
display(base2)
```

```python
agrupado = base2.groupby(['feed'])['weight'].sum() 
#Esse código agrupa os dados pela coluna "feed" e soma os valores da coluna "weight" para cada tipo de ração.
agrupado
```

```python
srn.kdeplot(base2.loc[base2['feed'] == 'horsebean','weight']).set(title='horsebean')
#USAR .SET(TITLE = ''), NAO SET_TITLE
```

```python
fig, axes = plt.subplots(3,2)

srn.histplot(base2.loc[base2['feed'] == 'casein','weight'], kde = True,ax=axes[0, 0]).set(title='casein')
srn.histplot(base2.loc[base2['feed'] == 'linseed','weight'], kde = True,ax=axes[0, 1]).set(title='linseed')
srn.histplot(base2.loc[base2['feed'] == 'meatmeal','weight'], kde = True, ax=axes[1, 0]).set(title='meatmeal')
srn.histplot(base2.loc[base2['feed'] == 'soybean','weight'], kde = True, ax=axes[1, 1]).set(title='soybean')
srn.histplot(base2.loc[base2['feed'] == 'sunflower','weight'], kde = True, ax= axes[2, 0]).set(title='sunflower')
srn.histplot(base2.loc[base2['feed'] == 'horsebean','weight'], kde = True,ax=axes[2, 1]).set(title='horsebean')


plt.tight_layout()

```
