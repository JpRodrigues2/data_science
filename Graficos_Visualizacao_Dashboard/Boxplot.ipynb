```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
# MATPLOTLIB

plt.boxplot(base.Volume, vert = False, showfliers = False, notch= True, patch_artist = True)
# VERT = boxplot gerado na vertical ou horizontal
# showfliers = mostar pontos outliers
# Notch = talho no centro
# patch_artist = preencher (cor) o boxplot

plt.title('Árvores')
plt.xlabel("Volume")
```

```python
srn.boxplot(data = base.Volume, orient = 'v').set_title("Árvores") #orient vetical ou horizontal
#PASSA COMO PARAMETROS O 'DATA' .x PARA ALGUMA COLUNA ESPECIFICA
```

```python
srn.boxplot(data = base, orient = 'h').set_title("boxplot's")
```
